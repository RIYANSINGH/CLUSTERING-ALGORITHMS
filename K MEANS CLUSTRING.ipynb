# IMPORTING THR DATA SET

here we are omly taking the 2 features of the data set ,that will make visulization easy . we have selected the columbs by index ie. [3,4] as shown below


```python
dataset = pd.read_csv(r"C:\Users\RIYAN SINGH\Desktop\course\Machine Learning A-Z (Codes and Datasets)\Part 4 - Clustering\Section 24 - K-Means Clustering\Python\Mall_Customers.csv")
x = dataset.iloc[:, [3,4]].values
```

here we are calculating the number of clusters we have to take for the daata by using the method of elbow 
in the we run the for loop up to 10 clusters and visulize the elbow by drawing the grapy 


```python
from sklearn.cluster import KMeans
wcss = []# sum of sqr distances of the observations
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

as you can see above that elbow is formed on cluster no 5 so its the obtained number of clusters required for the data set

# TRAINING THE MODEL

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans=kmeans.fit_predict(x)
```

```python
y_kmeans# you ncan se the the data is clustered into 5 classes 01234
```

# VISULIZATION THE PRIDICTIONS


```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='r'),#lable='CLUSTER 1')#here we are taking the 2paramert5es in the x k==0 means the value with the 0 class will be ploted 
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='b'),#lable='CLUSTER 2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='g'),#lable='CLUSTER 3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=100,c='orange')#,lable='CLUSTER 4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=100,c='grey'),#lable='CLUSTER 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='y',label='centeriode')
plt.title('CLUSTER OF CUSTUMERS')
plt.xlabel('ANUAL INCOME ')
plt.ylabel('SS')
plt.show()
```
